```python
import pandas as pd
```

```python
data = { 
    "Name": ["Alice", "Bob", "Charlie", "Diana", None],
    "Age": [25, None, 30, 22, 28],
    "City": ["New York", "Paris", None, "London", "Berlin"],
    "Income": [50000, 60000, None, 45000, 52000]  
}
```

```python
df = pd.DataFrame(data)
```

```python
print("Original Dataset:\n", df)
```

```python
print("Missing Values Summary:")
df.isnull().sum()
```

```python
# Drop rows with any missing value
clean_df = df.dropna()
print("After dropping missing values:\n", clean_df)
```

```python
# Filling missing values (imputation)
df.fillna({
    "Age": df["Age"].mean(),
    "Income": df["Income"].mean(),
    "City": df["City"].mode()[0],
    "Name": "Unknown"
}, inplace=True)
```

```python
# Advanced Imputation with scikit-learn
from sklearn.impute import SimpleImputer
import numpy as np

data = np.array([[1, 2, np.nan], [3, np.nan, 9], [7, 6, 5]])
imputer = SimpleImputer(strategy="mean")
filled = imputer.fit_transform(data)
print("After ML-based Imputation:\n", filled)
```
