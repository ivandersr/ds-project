```python
import pandas as pd
```

```python
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 28],
    "City": ["New York", "London", "Paris"]
}
```

```python
df = pd.DataFrame(data)
```

```python
# CSV
df.to_csv("../data/raw/people.csv", index=False)
loaded_df = pd.read_csv("../data/raw/people.csv")
loaded_df
```

```python
# XLSX
# Note: For workflows, it is usually recommended to convert Excel files to CSV or Parquet
df.to_excel("../data/raw/people.xlsx", index=False, sheet_name="Contacts")
excel_df = pd.read_excel("../data/raw/people.xlsx", sheet_name="Contacts")
excel_df
```

```python
# JSON
import json

data = {
    "employee":{
        "name": "alice",
        "department": "Analytics",
        "skills": ["Python", "R", "SQL"]
    }
}

with open("../data/raw/employee.json", "w") as f:
    json.dump(data, f, indent=4)

with open("../data/raw/employee.json", "r") as f:
    loaded_data = json.load(f)

flat_df = pd.json_normalize(data, sep="_")
flat_df
```

```python
# Parquet and Feather (Binary Columnar Formats)
df.to_parquet("../data/raw/people.parquet")
parquet_df = pd.read_parquet("../data/raw/people.parquet")
parquet_df
```

```python
# SQL
import sqlite3
conn = sqlite3.connect(":memory:")
df.to_sql("people", conn, index=False, if_exists="replace")
query_df = pd.read_sql("SELECT * FROM people WHERE Age > 25", conn)
query_df
```
