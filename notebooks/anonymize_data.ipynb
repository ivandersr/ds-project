```python
import pandas as pd
from faker import Faker
```

```python
data = {
    "user_id": [101, 102, 103],
    "name": ["Alice Johnson", "Bob Smith", "Charlie Brown"],
    "email": ["alice@example.com", "bob@example.com", "charlie@example.com"],
    "purchase_amount": [120.5, 89.9, 230.0]
}
```

```python
df = pd.DataFrame(data)
```

```python
fake = Faker()
df["name"] = [fake.name() for _ in df.index]
df["email"] = [fake.email() for _ in df.index]
```

```python
df.to_csv("../data/raw/anonymized_purchases.csv", index=False)
```

```python
from cryptography.fernet import Fernet
```

```python
key = Fernet.generate_key()
cipher = Fernet(key)
```

```python
original_data = "Sensitive_User_Info"

encrypted_data = cipher.encrypt(original_data.encode())
encrypted_data
```

```python
decrypted_data = cipher.decrypt(encrypted_data).decode()
decrypted_data
```
