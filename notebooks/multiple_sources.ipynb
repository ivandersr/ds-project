```python
import pandas as pd
import requests
```

```python
sales = pd.DataFrame({
    "Product": ["Laptop", "Phone", "Tablet"],
    "Revenue_USD": [150000, 90000, 36000]
})
```

```python
currency_rate_url = "https://api.fxratesapi.com/latest"
currency_rate_response = requests.get(currency_rate_url)
currency_rate_data = currency_rate_response.json()
```

```python
currency_rate_data
```

```python
usd_to_brl = currency_rate_data["rates"]["BRL"]
usd_to_brl
```

```python
sales["Revenue_BRL"] = (sales["Revenue_USD"] * usd_to_brl).round(2)
```
